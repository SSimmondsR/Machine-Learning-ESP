# ACTIVIDAD 1

**Limpieza de un carácter especial.**



```python
# Para la limpieza de datos en el dataframe se realiza una iteración para que recorra las columnas y verifique las columnas de tipo String para hacer esta limpieza
for col in df.columns:
  # Con el siguiente codigo se verifica si la columna es de tipo String para realizar la limpieza
  if df[col].dtype == 'object':
    # Se aplica el metodo de Strip para cada string en la columna
    df[col] = df[col].str.strip()
    # Se aplica limpieza de caracteres no ASCII
    df[col] = df[col].str.replace(r'[^\x00-\x7F]+', '', regex=True)
    # Se aplica limpieza de tabulaciones y saltos de linea
    df[col] = df[col].str.replace(r'\t|\n|\r', ' ', regex=True)

print(df)
```

**Considerando que algunas columnas presentan valores en PESOS y que posteriormente se desea realizar modelos predictivos o clustering, se debe eliminar el caracter $ de estas columnas para que sea tomado como una variable de tipo FLOAT.**

```python
# Lista de columnas a limpiar
columnas_a_limpiar = ['INGRESOS OPERACIONALES', 'GANANCIA (PÉRDIDA)', 'TOTAL ACTIVOS', 'TOTAL PASIVOS', 'TOTAL PATRIMONIO']

# Eliminar el carácter $ y convertir a float
df[columnas_a_limpiar] = df[columnas_a_limpiar].replace({'\$': ''}, regex=True)

print("\nDataFrame después de eliminar el carácter $:")
print(df)
```

**Reducción de los nits de longitud mayor a 9.**

Medicante la función `.str[:9].astype(int)` para obtener los primeros 9 caracteres de la cadena y se define el tipo de variable a entero.

```python
df['NIT'] = df['NIT'].astype(str).str[:9].astype(int)
print(df.NIT)
```

# ACTIVIDAD 2

 **Transformación del tipo de dato.**

 Mediante la función de `df.astype({'Columna': float}) ` se transforma la columna a una variable de tipo float, para este caso se selecciono la columna CIIU el cual contenia variables de tipo int y se hizo la transformacion a variable float.

```python
# Convirtiendo las columnas que contiene valores de dinero a float, anteriormente se eliminó el caracter $ de estas columnas definidas en columnas_a_limpiar.
df[columnas_a_limpiar].astype(float)
print(df.CIIU)
# Verificando el tipo de datos después de la conversión

print(df.dtypes)


```
