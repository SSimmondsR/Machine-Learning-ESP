# 1. Transformación del tipo de dato.

Se realiza la normalización de los datos importantes para realizar un posterior análisis de los datos, para ello se hace uso de la función `pd.to_numeric` y `pd.to_datatime `

```python
# Crear un diccionario para mapear los nombres de columnas de df_secop_1 a los de df_secop_2
rename_columns = {
    'identificador_paa': 'id_plan_anual_de_adquisiciones',
    'nombre_entidad': 'nombre_entidad',
    'nit_entidad': 'nit_entidad',
    'anno_paa': 'anno',
    'telefono_entidad': 'telefono_contacto',
    'mision_y_vision': 'mision_vision',
    'perspectiva_estrategica': 'perspectiva_estrategica',
    'ppto_total': 'valor_presupuesto_general',
    'limite_menor_cuantia': 'presupuesto_menor_cuantia',
    'limite_minima_cuantia': 'presupuesto_minima_cuantia',
    'fecha_ultima_actualizacion': 'fecha_de_ultima_modificacion',
    'fecha_cargue': 'fecha_de_creacion',
    'municipio_paa': 'municipio_paa',
    'departamento_paa': 'departamento_paa'
}

# Renombrar las columnas en df_secop_1
df_secop_1.rename(columns=rename_columns, inplace=True)

# Verificar los nuevos nombres de las columnas
print(df_secop_1.columns)

# Convertir columnas mediante la función pd.to_numeric la cual convierte una serie de datos (como una columna de un DataFrame) a un tipo numérico (int64, float64). Con el parametro errors='coerce' cualquier valor que no pueda ser convertido a un número se convertira en un valor NaN

df_secop_1['valor_presupuesto_general'] = pd.to_numeric(df_secop_1['valor_presupuesto_general'], errors='coerce')
df_secop_1['presupuesto_menor_cuantia'] = pd.to_numeric(df_secop_1['presupuesto_menor_cuantia'], errors='coerce')
df_secop_1['presupuesto_minima_cuantia'] = pd.to_numeric(df_secop_1['presupuesto_minima_cuantia'], errors='coerce')


df_secop_2['valor_presupuesto_general'] = pd.to_numeric(df_secop_2['valor_presupuesto_general'], errors='coerce')
df_secop_2['presupuesto_menor_cuantia'] = pd.to_numeric(df_secop_2['presupuesto_menor_cuantia'], errors='coerce')
df_secop_2['presupuesto_minima_cuantia'] = pd.to_numeric(df_secop_2['presupuesto_minima_cuantia'], errors='coerce')

# Convertir columnas de fechas al tipo datetime esta vez utilizando la función pd.to_datatime la cual reconoce fechas en variables de tipo strings y el parametro errors='coerce' convierte a NaN cualquier valor que no pueda reconocerse como fecha.
df_secop_1['fecha_de_ultima_modificacion'] = pd.to_datetime(df_secop_1['fecha_de_ultima_modificacion'], errors='coerce')
df_secop_1['fecha_de_creacion'] = pd.to_datetime(df_secop_1['fecha_de_creacion'], errors='coerce')

df_secop_2['fecha_de_publicacion'] = pd.to_datetime(df_secop_2['fecha_de_publicacion'], errors='coerce')
df_secop_2['fecha_de_creacion'] = pd.to_datetime(df_secop_2['fecha_de_creacion'], errors='coerce')
df_secop_2['fecha_de_ultima_modificacion'] = pd.to_datetime(df_secop_2['fecha_de_ultima_modificacion'], errors='coerce')

# Eliminar guiones "-" y espacios " " de la columna nit_entidad para df_secop_2

df_secop_2['nit_entidad'] = df_secop_2['nit_entidad'].str.replace('-', '').str.replace(' ', '')
df_secop_1['nit_entidad'] = df_secop_1['nit_entidad'].str.replace('-', '').str.replace(' ', '')

df_secop_1.info()
df_secop_2.info()
```

# 2. Cruce de tablas

```python

# Lista de columnas a seleccionar
columnas_seleccionadas = [
    'id_plan_anual_de_adquisiciones',
    'nombre_entidad',
    'nit_entidad',
    'anno',
    'telefono_contacto',
    'mision_vision',
    'perspectiva_estrategica',
    'valor_presupuesto_general',
    'presupuesto_menor_cuantia',
    'presupuesto_minima_cuantia',
    'fecha_de_ultima_modificacion',
    'fecha_de_creacion',
    'municipio_paa',
    'departamento_paa'
]

# Seleccionar las columnas en df_secop_1 y df_secop_2
df_secop_1_seleccionado = df_secop_1[columnas_seleccionadas]
df_secop_2_seleccionado = df_secop_2[columnas_seleccionadas]

# Concatenar los DataFrames por filas
df_concatenado = pd.concat([df_secop_1_seleccionado, df_secop_2_seleccionado], axis=0, ignore_index=True)

# Verificar los primeros resultados del DataFrame concatenado
print(df_concatenado.head())
```

# 3. Filtro de variables por algún tipo de criterio

Se realizan 2 filtros, el primero de ellos filtra todos los valores del 2021 y la ciudad de Bogotá el el df concatenado, el otro filtro toma los registros del 2025 para el departamento de Antioquia.

```python
# Filtrar por el año 2023 en ambos DataFrames
df_filtered_Bogota_2021 = df_concatenado[(df_concatenado['anno'] == 2021) & (df_concatenado['municipio_paa'] == 'Bogotá')]
df_filtered_antioquia_2024 = df_concatenado[(df_concatenado['anno'] == 2024) & (df_concatenado['departamento_paa'] == 'Antioquia')]
df_filtered_Bogota_2021.head()
df_filtered_antioquia_2024.head()
```

#  Actividad 2. Visualización



```python
# Comparación del número de registros por año
plt.figure(figsize=(12, 6))
sns.countplot(x='anno', data=df_concatenado, palette='viridis')
plt.title('Número de Registros por Año')
plt.xlabel('Año')
plt.ylabel('Cantidad de Registros')
print(f"Grafica 1.")
plt.show()

# Comparar la suma total de valor presupuesto general por departamento
plt.figure(figsize=(12, 6))
df_grouped = df_concatenado.groupby('departamento_paa')['valor_presupuesto_general'].sum().reset_index()
sns.barplot(x='departamento_paa', y='valor_presupuesto_general', data=df_grouped, palette='coolwarm')
plt.title('Suma Total del Valor Presupuesto General por Departamento')
plt.xlabel('Departamento')
plt.ylabel('Suma Total del Valor Presupuesto General')
plt.xticks(rotation=90)
print(f"Grafica 2.")
plt.show()
```

### A partir de las graficas obtenidas podemos concluir:


*   En el año 2024 se realizaron mayor cantidad de registros en el SECOP, seguido por el año 2022 - 2021 - 2024 respecticamente. (Grafica 1)
*   Antioquia fue el departamento con mayor valor presupuestal general, seguido de Atlántico y Distrito Capital de Bogotá respecticamente. (Grafica 2)
