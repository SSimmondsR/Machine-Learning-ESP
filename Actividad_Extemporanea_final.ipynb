# datos de entrada:

***Base de datos 2 - Alcohol en estudiantes***

**school** - student's school (binary: 'GP' - Gabriel Pereira or 'MS' - Mousinho da Silveira)

**sex** - student's sex (binary: 'F' - female or 'M' - male)

**age** - student's age (numeric: from 15 to 22)

**address** - student's home address type (binary: 'U' - urban or 'R' - rural)

**famsize** - family size (binary: 'LE3' - less or equal to 3 or 'GT3' - greater than 3)

**Pstatus** - parent's cohabitation status (binary: 'T' - living together or 'A' - apart)

**Medu** - mother's education (numeric: 0 - none, 1 - primary education (4th grade)
, 2 – 5th to 9th grade, 3 – secondary education or 4 – higher education)

**Fedu** - father's education (numeric: 0 - none, 1 - primary education (4th grade), 2 – 5th to 9th grade, 3 – secondary education or 4 – higher education)

**Mjob** - mother's job (nominal: 'teacher', 'health' care related, civil 'services' (e.g. administrative or police), 'at_home' or 'other')

**Fjob** - father's job (nominal: 'teacher', 'health' care related, civil 'services' (e.g. administrative or police), 'at_home' or 'other')

**reason** - reason to choose this school (nominal: close to 'home', school 'reputation', 'course' preference or 'other')

**guardian** - student's guardian (nominal: 'mother', 'father' or 'other')

**traveltime** - home to school travel time (numeric: 1 - <15 min., 2 - 15 to 30 min., 3 - 30 min. to 1 hour, or 4 - >1 hour)

**studytime** - weekly study time (numeric: 1 - <2 hours, 2 - 2 to 5 hours, 3 - 5 to 10 hours, or 4 - >10 hours)

**failures** - number of past class failures (numeric: n if 1<=n<3, else 4)

**schoolsup** - extra educational support (binary: yes or no)

**famsup** - family educational support (binary: yes or no)

**paid** - extra paid classes within the course subject (Math or Portuguese)
(binary: yes or no)

**activities** - extra-curricular activities (binary: yes or no)

**nursery** - attended nursery school (binary: yes or no)

**higher** - wants to take higher education (binary: yes or no)

**internet** - Internet access at home (binary: yes or no)

**romantic** - with a romantic relationship (binary: yes or no)

**famrel** - quality of family relationships (numeric: from 1 - very bad to 5 - excellent)

**freetime** - free time after school (numeric: from 1 - very low to 5 - very high)

**goout** - going out with friends (numeric: from 1 - very low to 5 - very high)

**Dalc** - workday alcohol consumption (numeric: from 1 - very low to 5 - very high)

**Walc** - weekend alcohol consumption (numeric: from 1 - very low to 5 - very high)

**health** - current health status (numeric: from 1 - very bad to 5 - very good)

**absences** - number of school absences (numeric: from 0 to 93)

These grades are related with the course subject, Math or Portuguese:

**G1 - first period grade (numeric: from 0 to 20)**

**G2 - second period grade (numeric: from 0 to 20)**

**G3 - final grade (numeric: from 0 to 20, output target)**

# **4.    Análisis exploratorio de las variables.**

```python
# Creación de los dataframes
df_student_mat = pd.read_csv('student-mat.csv', sep = ',')
df_student_por = pd.read_csv('student-por.csv', sep = ',')

# Información general de los dataframes
print(f"Tipo de variables en base de datos estudiante de curso de  matematicas \n",  df_student_mat.info())
print(f" Tipo de variables en base de datos estudiantes de  curso de portugues \n" ,df_student_por.info())

# Ver las primeras 5 filas en ambos dataframes
print(f"Primeras 5 filas de la base de datos estudiantes de matematicas \n" , df_student_mat.head())
print(f"Primeras 5 filas de la base de datos estudiantes de portugues \n" , df_student_por.head())

# Resumen estadístico de los dataframes
print(f" Resumen estadistico base de datos curso de matematicas \n" , df_student_mat.describe())
print(f" Resumen estadistico base de datos curso de portugues \n" , df_student_por.describe())

# Verificación de valores nulos
print(f"\n Valores nulos en base de datos estudiantes de matemáticas \n" , df_student_mat.isnull().sum())
print(f"\n Valores nulos en base de datos estudiantes de portugues \n" , df_student_por.isnull().sum())
```

5.    Cruce de bases.

```python
# Concatenar los DataFrames estudiantes clase de matemáticos y estudiantes clase de portugues en un nuevo DF, se debe considerar que se encuentran registrados 382 estudiantes en ambos cursos (Matematicas y Portugues)
df_concat = pd.concat([df_student_mat, df_student_por], ignore_index=True)
df_concat.info()
```

# **6.    Filtro de bases.**

```python
# Estadisticas calificaciones generales
print(f"Estadistica calificaciones generales \n", df_concat[['G1','G2','G3']].describe())

# Filtrar estudiantes con bajo consumo de alcohol durante la semana (Dalc <= 2)
df_low_dalc = df_concat[df_concat['Dalc'] <= 2]
print(f"Análisis CALIFICACIONES estudiantes con bajo consumo de alcohol durante la semana (Dalc <= 2): \n", df_low_dalc[['G1','G2','G3']].describe())

# Filtrar estudiantes con alto consumo de alcohol durante la semana (Dalc >= 4)
df_high_dalc = df_concat[df_concat['Dalc'] >= 4]
print(f"Análisis CALIFICACIONES estudiantes con alto consumo de alcohol durante la semana (Dalc >= 4): \n", df_high_dalc[['G1','G2','G3']].describe())

# Filtrar estudiantes con alto consumo de alcohol durante el fin de semana (Walc >= 4)
df_high_walc = df_concat[df_concat['Walc'] >= 4]
print(f"Análisis CALIFICACIONES estudiantes con alto consumo de alcohol durante el fin de semana (Walc >= 4): \n", df_high_walc[['G1','G2','G3']].describe())

# Filtrar estudiantes con bajo consumo de alcohol durante el fin de semana (Walc =< 2)
df_low_walc = df_concat[df_concat['Walc'] <= 2]
print(f"Análisis CALIFICACIONES estudiantes con bajo consumo de alcohol durante el fin de semana (Walc <= 2): \n", df_low_walc[['G1','G2','G3']].describe())

# Filtrar estudiantes con padres JUNTOS (Pstatus = 'T')
df_p_t = df_concat[df_concat['Pstatus'] == 'T']
print(f"Análisis resultados académicos de estudiantes con padres juntos (Pstatus = T): \n", df_p_t[['G1','G2','G3']].describe())

# Filtrar estudiantes con padres SEPARADOS (Pstatus = 'A')
df_p_a = df_concat[df_concat['Pstatus'] == 'A']
print(f"Análisis resultados académicos de estudiantes con padres separados (Pstatus = A): \n", df_p_a[['G1','G2','G3']].describe())
```



*   El filtro de estudiantes y el consumo de alcohol durante semana y fines de semana y la descripción de este filtro, nos acerca a los primeros análisis que se pueden realizar para determinar el motivo de consumo de alcohol. Por ejemplo:

1. Como media general en las calificaciones se tiene notas de los 2 periodos y nota final promedio respectivamente de:  G1 11.22    G2 11.24    G3 11.34

2. Considerando que las calificaciones se tiene en un rango de (0 a 20) y que 20 corresponde al máximo resultado posible, para los estudiantes que tienen alto consumo de alcohol los fines de semana se tienen muy bajas calificaciones en los dos periodos y en su nota final (calificacion de 10.4 como media), al igual que los estudiantes con bajo consumo de alcohol los fines de semana, por lo que esta variable de notas no determinaría un motivo de consumo de alcohol.

3. Los estudiantes que tienen alto consumo de alcohol durante la semana, obteniendo peores calificaciones que los estudiantes que tienen bajo consumo de alcohol durante la semana, este parametro indica que los estudiantes ven afectados el rendimiento en los cursos por el alto consumo de alcohol entre semana. (de 11.4 a 9.82 de media en la nota final).

4. Las estadísticas presentan mejores resultados académicos en estudiantes con padres separados que con padres juntos (Nota final 11,66 PS y 11,29 PJ).









# **7.    Graficación de las variables.**

```python
# Configurar el estilo de los gráficos
sns.set(style="whitegrid")

# Crear un DataFrame para las barras agrupadas
df_melted = pd.melt(df_concat, id_vars='sex', value_vars=['Dalc', 'Walc'],
                    var_name='Consumo_Alcohol', value_name='Level')

# Graficar la relación
plt.figure(figsize=(10, 6))
sns.barplot(x='Consumo_Alcohol', y='Level', hue='sex', data=df_melted, ci=None)

# Añadir título y etiquetas

plt.title('Consumo de Alcohol entre Semana (Dalc) vs Fin de Semana (Walc) por Sexo')
plt.xlabel('Contexto de Consumo de Alcohol')
plt.ylabel('Nivel de Consumo de Alcohol')
plt.legend(title='Sexo')

# Mostrar la gráfica
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(df_concat['Walc'], bins=5, kde=True)
plt.title('Distribución del consumo de alcohol en fines de semana')
plt.xlabel('Walc (Consumo de alcohol en fines de semana)')
plt.ylabel('Frecuencia')
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(x='school', y='Dalc', data=df_concat)
plt.title('Promedio de consumo de alcohol en días laborables por colegio')
plt.xlabel('Colegio')
plt.ylabel('Promedio de Dalc')
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x='sex', y='G3', data=df_concat)
plt.title('Distribución de calificaciones (G3) por sexo')
plt.xlabel('Sexo')
plt.ylabel('Calificación final (G3)')
plt.show()


# Crear una figura con dos subplots en una fila
fig, axes = plt.subplots(1, 2, figsize=(16, 5))  # 1 fila, 2 columnas
fig.suptitle('Resultado programas de apoyo frente a consumo de alcohol', size=14, weight=25, color='r')

# Gráfico para schoolsup vs Dalc
sns.boxplot(x='schoolsup', y='Dalc', data=df_concat, ax=axes[0])
axes[0].set_title('Consumo de alcohol en días laborables vs Apoyo Escolar')
axes[0].set_xlabel('Apoyo Escolar')
axes[0].set_ylabel('Consumo de alcohol en días laborables')

# Gráfico para famsup vs Dalc
sns.boxplot(x='famsup', y='Dalc', data=df_concat, ax=axes[1])
axes[1].set_title('Consumo de alcohol en días laborables vs Apoyo Familiar')
axes[1].set_xlabel('Apoyo Familiar ')
axes[1].set_ylabel('Consumo de alcohol en días laborables')
```
