````python
import pandas as pd

df = pd.read_csv('census_income.csv')
print(df)
````

Hacemos una limpieza y ajuste de la base de datos para pasar a trabajar sobre esta.
Luego de la limpieza y con la funcion .info() obtenemos la informacion de las columnas y su tipo de dato.



````python
# Debido a que el archivo se encuentra en 1 sola columna separado por ';' hacemos la separacion de las columnas
df = pd.read_csv('census_income.csv', delimiter=';')

# Verificamos la estructura de nuevo donde se presentan la cantidad de datos por columnas, y tipo de dato.
df_info = df.info()
df_head = df.head()

df_info, df_head
````

Al revisar el tipo de variable que define `object` generalmente representa cadenas de texto (strings), sin embargo, puede contener otros tipos de datos como listas o diccionarios.

El otro tipo de variable que tenemos en esta base de datos es de tipo numerico `int` (entero).

Una vez obtenido el tipo de datos que se encuentra en la base de datos, mediante la funcion de DataFrame de la libreria Pandas se puede realizar una tabla que incluya las columnas deseadas, en este caso las columnas seran la Cantidad de valores, Valores distintos y Valores Nulos de los diferentes atributos de nuestra base de datos.


````python
# Para cada columna, obtenemos:# - Cantidad de valores# - Cantidad de valores distintos# - Cantidad de valores nulos

resumen_de_columnas = pd.DataFrame({
    'Cantidad de valores': df.count(),
    'Valores distintos': df.nunique(),
    'Valores nulos': df.isnull().sum()
})

resumen_de_columnas
````
